```
# This block concats normal packets and malicious packets
import os
import pandas as pd
path = "drive/My Drive/Assessment5_DST/"
os.chdir(path)
normalPath = "Data/benign"
malPath = "Data/malicious"
bFrame = pd.DataFrame()
mFrame = pd.DataFrame()
for file in os.listdir(normalPath): # Read all normal data and concat them
  tempdf = pd.read_csv(normalPath+"/"+file)
  bFrame = pd.concat([bFrame, tempdf])
bFrame["type"] = "benign" 

for file in os.listdir(malPath): # Read all malicious data
  tempdf = pd.read_csv(malPath+"/"+file)
  tempdf["type"] = file.split(".")[0]
  mFrame = pd.concat([mFrame, tempdf])

mFrame.head(5)
```

```
len(mFrame), len(bFrame)
```

```
from sklearn.utils import shuffle
bf = bFrame.sample(n = 50000)
mf = mFrame.sample(n = 250000) # Randomly choose 50000 data from benign and 250000 data from malicious packets
concat_df = pd.concat([bf, mf])
concat_df = shuffle(concat_df) # Shuffle the index

concat_df.head(5)
```

```
attacks = ["combo", "junk", "scan", "tcp", "udp"]
for tp in attacks:
  print("The number of ", tp, " packets are: ", concat_df[concat_df["type"] ==  tp].count()[0])
```

```
# Get the column names and prepare for spark
col_names = concat_df.columns
```

```
concat_df.to_csv("Data/concat/concat_file.csv", header = True, index = False) # Write out concat file
```

```
f = open("Data/concat/columns.txt", "w") # Write out column names
for col in col_names:
  f.write(col)
  f.write("\n")
f.close()
```
